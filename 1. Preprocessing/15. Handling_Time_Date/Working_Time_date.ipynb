# Working with dates

```python
date['date']=pd.to_datetime(date['date'])
```

```python
date.info()
```

## Extract year

```python
date['Date_Year']=date['date'].dt.year

date.head()
```

## Extract Month

```python
date['Date_Month_no']=date['date'].dt.month
date.head()
```

```python
date['Date_Month_name']=date['date'].dt.month_name()
date.head()
```

```python
date.drop(['product_id','city_id','orders'],axis=1,inplace=True)
```

```python
date.head()
```

## Extract Date

```python
date['date_day']=date['date'].dt.day
date.head()
```

### day of a week

```python
date['date_dow']=date['date'].dt.dayofweek
date.head()
```

```python
date['date_dow_name']=date['date'].dt.day_name()
date.head()
```

```python
date['is_weeknd']=np.where(date['date_dow_name'].isin(['Sunday','Saturday']),1,0)
date.sample(5)
```

### Extract Week of the Year

```python
date['Week_no_year']=date['date'].dt.week
date.head()
```

```python
date['quarter']=date['date'].dt.quarter
date.sample(5)
```

```python
date['semester']=np.where(date['quarter'].isin([1,2]),1,2)
date.sample(5)
```

```python
import datetime

today=datetime.datetime.today()
today
```

```python
today-date['date']
```

```python
(today-date['date']).dt.days
```

```python
#month passed
np.round((today -date['date']) / np.timedelta64(1, 'M'),0)
```
