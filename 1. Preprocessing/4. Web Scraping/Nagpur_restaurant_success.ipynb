```python
import pandas as pd
import numpy as np
import requests 
from bs4 import BeautifulSoup
```

```python
final=pd.DataFrame()
page=1
for i in range(1,40):
    print("Page: ",page)
    page+=1
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    url=f"https://www.dineout.co.in/nagpur-restaurants?p={i}"
    webpage=requests.get(url,headers=headers).text
    
    soup=BeautifulSoup(webpage,'lxml')
    rating_class=soup.select(f'[class*="restnt-rating rating-"]')
    
    hideCount=1
    rating=[]
    zone=[]
    area=[]
    name=[]
    rate=[]
    cuisine=[]
    
    for i in rating_class:
        class_selector=i['class']

        if hideCount>1:
            rating.append(np.nan)
            hideCount-=1


        if 'hide' in class_selector:
            hideCount+=1
        else:
            rating.append(i.text)
            hideCount=0

    if hideCount>1:
        rating.append(np.nan)
        hideCount-=1
    
    for j in soup.find_all('div',class_="restnt-loc ellipsis"):
        a=j.text.strip().split(",")
        if(len(a)>2):
            ar=a[0]+","+a[1]
            try:
                area.append(ar)
            except:
                area.append(np.nan)
            try:
                zone.append(a[-1])
            except:
                zone.append(np.nan)
        else:
            try:
                area.append(a[0])
            except:
                area.append(np.nan)
            try:
                zone.append(a[-1])
            except:
                zone.append(np.nan)
            
            
    for j in soup.find_all('a',class_='restnt-name ellipsis'):
        name.append(j.text)
    
    for j in soup.find_all('span',class_="double-line-ellipsis"):
        a=j.text.split('|')
        try:
            rate.append(a[0])
        except:
            rate.append(np.nan)
        try:
            cuisine.append(a[1])
        except:
            cuisine.append(np.nan)
    
    dic={
        "Name":name,
        'Rating':rating,
        "Cuisine":cuisine,
        "Avg Rate":rate,
        "Area":area,
        "Zone":zone
    }
    df=pd.DataFrame(dic)
    final=final.append(df,ignore_index=True)
    
    
    
```

```python
final.to_csv("Nagpur_restaurant.csv")
```
