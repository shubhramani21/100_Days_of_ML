```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("mental_health_finaldata_1.csv")
df.head()
```

```python
for i in df.columns:
    print(i, df[i].unique())
```

```python
df['Age'].value_counts()
```

```python
fig,ax=plt.subplot(nrows=(len(df.columns)/2),ncols=(len(df.columns)/2),figsize=())
for i in df.columns:  
    count=df[i].value_counts()
    plt.figure(figsize=(8,6))
    count.plot(kind='bar',color='skyblue')
    plt.title(f'Distribution of {i}')
    plt.xlabel(i)
    plt.ylabel('Count')
    plt.show()
```

```python


# Assuming your data is in a DataFrame df
fig, axes = plt.subplots(nrows=len(df.columns), ncols=1, figsize=(8, 2* len(df.columns)))

for i, column in enumerate(df.columns):
    counts = df[column].value_counts()
    axes[i].pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=140)
    axes[i].set_title(f'Distribution of {column}')

plt.tight_layout()
plt.show()
```
