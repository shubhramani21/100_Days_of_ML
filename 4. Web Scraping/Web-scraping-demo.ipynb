```python
final=pd.DataFrame()
for j in range(1,21):
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    url=f"https://www.ambitionbox.com/list-of-companies?page={j}"
    webpage=requests.get(url,headers=headers).text
    soup=BeautifulSoup(webpage,'lxml')  
    
    name=[]
    rating=[]
    service=[]
    employees=[]
    sector=[]
    old=[]
    city=[]

    temp=[]

    for i in soup.find_all('h2',class_="companyCardWrapper__companyName"):
        name.append(i.text.strip())
        
        
    for i in soup.find_all('span',class_="companyCardWrapper__companyRatingValue"):
        rating.append(float(i.text.strip()))
        
        
    for i in soup.find_all('span',class_="companyCardWrapper__interLinking"):
        temp=i.text.strip().split('|');
        if(len(temp)==5 or len(temp)==4):
            service.append(temp[0])
            employees.append(temp[1])
            old.append(temp[-2])
            city.append(temp[-1].split('+')[0].strip())
    
    dic={
        "name":name,
        "rating":rating,
        "service":service,
        "employees":employees,
        "since":old,
        "location":city
    }
    df=pd.DataFrame(dic)
    
    final=final.append(df)
        
    
    
```
